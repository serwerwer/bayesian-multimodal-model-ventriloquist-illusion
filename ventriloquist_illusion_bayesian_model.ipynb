### Transforming data: we only want pairs of (AV, A) trials.

```python
# Function which creates a list of pairs where 'modality' == 0 is followed by 'modality' == 1
# 0 - audio-visual stimuli
# 1 - visual stimuli
# so each pair corresponds to one observation in which audio-visual stimuli is followed by visual one

def modality_pairs(df):
    pairs = []
    i = 0
    while i < len(df) - 1:
        if df.loc[i, 'modality'] == 0:

            # Skipping the first 0 if followed by another 0
            if i < len(df) - 2 and df.loc[i + 1, 'modality'] == 0:
                i += 1

            # Going to the next row if the participant is changing between consecutive rows
            elif df.loc[i, 'SubId'] != df.loc[i + 1, 'SubId']:
                i += 1

            # If we have a pair (0, 1)
            elif df.loc[i + 1, 'modality'] == 1:
                pairs.append([df.iloc[i], df.iloc[i + 1]])
                i += 2

            # Going to the next row if the next modality value is 0 for the same participant
            else:
                i += 1

        # Going to the next row if participant is changing
        else:
            if df.loc[i, 'SubId'] != df.loc[i + 1, 'SubId']:
                i += 1

        # Moving to the next row
        i += 1

    return pairs

pairs = modality_pairs(df)
pairs
```

```python
pairs_combined = []

for pair in pairs:
    pair_dict = {}
    for idx, row in enumerate(pair):
        for col_name, value in row.items():
            new_col_name = f'{col_name}_A' if idx != 0 else col_name
            pair_dict[new_col_name] = value
    pairs_combined.append(pair_dict)

# Creating a new dataframe from the list of dictionaries
df_combined = pd.DataFrame(pairs_combined)

df_combined
```

## Models:
 *  Model 1 models the predicted response in the A trial given the response in the AV trial
 *  Model 2 models the predicted response in the A trial given the response in the AV trial, also taking into account the response in the previous A-AV trial pair.

```python
# Use standardized data?
# For now yes
relative_response_std, mean_relative_response, std_relative_response  = standardize(df_cleaned['Relative_response'], return_mean_and_std = True)
relative_response_A_std, mean_relative_response_A, std_relative_response_A  = standardize(df_cleaned['Relative_responses_A'], return_mean_and_std = True)

```

### Model 1

```python
import pymc as pm


# Model 1:
with pm.Model() as base_model_2:
    # Priors
    alpha = pm.Normal('alpha', mu=0, sigma=0.1)
    beta_resp = pm.Normal('beta_resp', mu=0, sigma=1)
    #beta_delta = pm.Normal('beta_delta', mu=0, sigma=1)

    sigma = pm.HalfCauchy('sigma', beta=5)

    # Multimodal prior receptivity
    #theta = pm.Normal('theta', mu=0, sigma=1)

    # Likelihood
    mu = alpha + beta_resp * relative_response_std #+ beta_delta * df_cleaned['DeltaAV']
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed= relative_response_A_std)

    # Sampling
    trace_base = pm.sample(return_inferencedata=True, idata_kwargs={"log_likelihood": True})
```

```python
delta_av_std = standardize(df_cleaned['DeltaAV'])
```

```python
import pymc as pm


# Model 1:
with pm.Model() as base_model_2:
    # Priors
    alpha = pm.Normal('alpha', mu=0, sigma=0.1)
    beta_resp = pm.Normal('beta_resp', mu=0, sigma=1)
    beta_shift = pm.Normal('beta_shift', mu=0, sigma=1)
    beta_delta = pm.Normal('beta_delta', mu=0, sigma=1)

    sigma = pm.HalfCauchy('sigma', beta=5)

    # Multimodal prior receptivity
    #theta = pm.Normal('theta', mu=0, sigma=1)

    # Likelihood
    mu = alpha + beta_resp * relative_response_std + beta_delta * delta_av_std + beta_shift * delta_av_std * relative_response_std
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed= relative_response_A_std)

    # Sampling
    trace_base_2 = pm.sample(return_inferencedata=True, idata_kwargs={"log_likelihood": True})
```

```python
# Model 2: Multimodal prior


with pm.Model() as multimodal_model:
    # Priors
    alpha = pm.Normal('alpha', mu=0, sigma=0.1)
    beta_resp = pm.Normal('beta_resp', mu=0, sigma=1)
    beta_prev = pm.Normal('beta_prev', mu=0, sigma=1)

    sigma = pm.HalfCauchy('sigma', beta=5)

    # Multimodal prior receptivity
    #theta = pm.Normal('theta', mu=0, sigma=1, shape=n_participants)

    # Likelihood
    mu = alpha + beta_resp * relative_response_std + beta_prev * previous_relative_response_std
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed= relative_response_A_std)

    # Sampling
    trace_multimodal = pm.sample(return_inferencedata=True, idata_kwargs={"log_likelihood": True})
```

```python
def plot_regression(trace, beta = "beta_prev"):
    plt.figure()
    # Extract posterior samples for alpha and beta
    alpha_samples = trace.posterior['alpha'].values.flatten()
    beta_samples = trace.posterior[beta].values.flatten()

    # Generate X values for plotting
    X_plot = np.linspace(-3, 3, 100)

    '''# Plot observed data
    plt.figure(figsize=(10, 6))
    plt.plot(X_plot, y, 'o', label="Observed data")
    '''
    # Plot sampled regression lines
    for a, b in zip(alpha_samples[::100], beta_samples[::100]):  # Plot a subset of lines for clarity
        plt.plot(X_plot, a + b * X_plot, color='black', alpha=0.1)


    predictions = np.array([a + b * X_plot for a, b in zip(alpha_samples, beta_samples)])
    hdi_cred = az.hdi(predictions, hdi_prob=0.95)

    # Plot mean regression line
    mean_regression_line = np.mean(predictions, axis=0)
    plt.plot(X_plot, mean_regression_line, color='blue', label=f'Mean regression line = {round(np.mean(beta_samples), 3)}')

    # Plot y = 0 line
    #plt.plot(X_plot, np.zeros_like(X_plot), color='red', label='y = 0 line', alpha = 0.5)

    # Plot HDI
    plt.fill_between(X_plot, hdi_cred[:, 0], hdi_cred[:, 1], color='gray', alpha=0.3, label='95% HDI')



    # Labels and legend
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title(f"Sampled Regression Lines, beta = {beta}")
    plt.legend()
    plt.show()

plot_regression(trace_multimodal, beta = "beta_prev")

plot_regression(trace_multimodal, beta = "beta_resp")

```
